```python
print("=== Expense Classifier ===")

amount = float(input("Enter an expense amount: $"))

if amount >= 100:
    print("Major Expense - consider your spending carefully")
elif amount >= 50:
    print("Moderate Expense - keep track of your budget")
elif amount >= 10:
    print("Minor Expense")
else:
    print("Small Purchase - no big impact")
```

```python
print("=== Budget Limit Checker ===")

monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter expense amount: $"))

ratio = expense / monthly_budget

if ratio > 0.5:
    print("Major purchase - think carefully")
elif ratio > 0.25:
    print("Significant expense")
elif ratio > 0.10:
    print("Moderate expense")
else:
    print("Minor expense - within budget")
```

```python
print("\n=== Savings Goal Tracker ===")

goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))

percentage = (current_savings / goal_amount) * 100

print(f"You have saved {percentage:.2f}% of your goal")

if percentage >= 100:
    print("Goal achieved! Consider a new goal.")
elif percentage >= 75:
    print("Almost there! You're doing great.")
elif percentage >= 50:
    print("Halfway there - keep it up!")
elif percentage >= 25:
    print("Good progress, stay consistent")
else:
    print("Just getting started - you've got this!")
```

```python
print("=== Smart Purchase Advisor ===")

available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Check if they can afford it
if purchase_price > available_money:
    needed = purchase_price - available_money
    print(f"You can't afford this. You need ${needed:.2f} more.")
else:
    ratio = purchase_price / available_money

    if ratio > 0.5:
        print("You can afford it, but it's a large portion of your money. Consider waiting.")
    elif ratio > 0.2:
        print("This is a reasonable purchase, but think about your budget.")
    else:
        print("This is a small purchase - go for it!")
```
