Goal: Create a simple program that sets up a user profile for a finance tracker.
The program will ask the user for a few details and then display a message using that information.

Data to collect:
- Name
- Theme color
- Currency
- Savings goal

Example:

Name: Tanya  
Theme color: dark  
Currency: USD  
Savings goal: a car  

Expected output:
Welcome to your Personal Finance Tracker, Tanya! Your profile is ready.
The app will use a dark theme and all amounts will be in USD.
Let's start working towards your goal of saving for a car!

1. Start the program  
2. Ask the user for their name  
3. Ask for their theme color  
4. Ask for their currency  
5. Ask for their savings goal  
6. Store all the answers  
7. Create a message using the inputs  
8. Display the message  
9. End the program  

```python
# Collect user information
user_name = input("What is your name? ")
theme_color = input("What color would you like for your app theme? ")
primary_currency = input("What is your primary currency? ")
main_savings_goal = input("What is your main savings goal? ")

# Create the message
summary = "Welcome to your Personal Finance Tracker, " + user_name + "! "
summary += "Your profile is ready. The app will use a " + theme_color
summary += " theme and all amounts will be in " + primary_currency + ". "
summary += "Let's start working towards your goal of saving for " + main_savings_goal + "!"

# Display the message
print(summary)
```
