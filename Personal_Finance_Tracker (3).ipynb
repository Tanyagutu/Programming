# Week 4 – Loop-Powered Finance Tracker

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print()

name = input("Enter your name: ")

# Budget validation
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        monthly_budget = float(input("Enter monthly budget ($100-$10,000): $"))

        if monthly_budget < 100:
            print("Budget too low!")
        elif monthly_budget > 10000:
            print("Budget too high!")
    except ValueError:
        print("Enter a valid number!")

print(f"\nWelcome {name}! Budget: ${monthly_budget:.2f}")

# ===== IMPORTANT PART =====
expenses = []
descriptions = []
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' to finish")

while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    try:
        amount = float(expense_input)

        if amount <= 0:
            print("Enter a positive number!")
            continue

        description = input("What was this expense for? ")

        expenses.append(amount)
        descriptions.append(description)

        total_spent += amount

        percent = (total_spent / monthly_budget) * 100

        print(f"Added: ${amount:.2f} for {description}")
        print(f"Total: ${total_spent:.2f} ({percent:.1f}%)")

    except ValueError:
        print("Enter a valid number!")

print(f"\nDone! {len(expenses)} expenses, ${total_spent:.2f} total")
```

```python
# Expense Analysis Section

if len(expenses) > 0:
    print("\n📊 Expense Analysis:")

    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent/len(expenses):.2f}")
    print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # Highest and lowest
    highest_expense = max(expenses)
    lowest_expense = min(expenses)

    print(f"\n💰 Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # Categorization
    large_count = 0
    medium_count = 0
    small_count = 0

    for expense in expenses:
        percent = (expense / monthly_budget) * 100

        if percent > 20:
            large_count += 1
        elif percent >= 5:
            medium_count += 1
        else:
            small_count += 1

    print(f"\n📊 Expense Categories:")
    print(f"Large expenses (>20%): {large_count}")
    print(f"Medium expenses (5–20%): {medium_count}")
    print(f"Small expenses (<5%): {small_count}")

    # Feedback
    print("\n💡 Smart Feedback:")

    if total_spent > monthly_budget:
        print("You have exceeded your budget!")
    elif total_spent > monthly_budget * 0.8:
        print("Warning: You are close to your budget limit.")
    else:
        print("Your spending is within a safe range.")

else:
    print("\nNo expenses entered.")
```

```python
print("\n=== Savings Goals Tracker ===")

goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals
num_goals = int(input("How many savings goals? (1-3): "))

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    name = input("Goal name: ")
    target = float(input("Target amount: $"))
    current = float(input("Current savings: $"))

    goal_names.append(name)
    goal_targets.append(target)
    goal_current.append(current)

print("\n📊 Goals Progress:")

for i in range(len(goal_names)):
    progress = (goal_current[i] / goal_targets[i]) * 100
    remaining = goal_targets[i] - goal_current[i]

    print(f"{goal_names[i]}: {progress:.1f}% complete")
    print(f"Remaining: ${remaining:.2f}")
```

```python
print("\n=== Interactive Finance Menu ===")

running = True

while running:
    print("\n" + "-"*30)
    print(f"{name}'s Finance Tracker")
    print(f"Expenses: {len(expenses)} | Total: ${total_spent:.2f}")
    print("-"*30)

    print("1. View expenses")
    print("2. View savings goals")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    # OPTION 1
    if choice == "1":
        print("\n--- Expense Summary ---")
        for i in range(len(expenses)):
            print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")

        print(f"Total: ${total_spent:.2f}")

    # OPTION 2
    elif choice == "2":
        print("\n--- Savings Goals ---")
        for i in range(len(goal_names)):
            progress = (goal_current[i] / goal_targets[i]) * 100
            print(f"{goal_names[i]}: {progress:.1f}% complete")

    # OPTION 3
    elif choice == "3":
        print("\n--- Add Expense ---")

        try:
            amount = float(input("Enter amount: $"))
            desc = input("Description: ")

            expenses.append(amount)
            descriptions.append(desc)
            total_spent += amount

            print("Expense added!")

        except ValueError:
            print("Invalid input!")

    # OPTION 4
    elif choice == "4":
        print("\n--- Budget Check ---")

        percent = (total_spent / monthly_budget) * 100
        remaining = monthly_budget - total_spent

        print(f"Budget used: {percent:.1f}%")
        print(f"Remaining: ${remaining:.2f}")

    # OPTION 5
    elif choice == "5":
        running = False
        print("\nGoodbye!")

    else:
        print("Invalid choice!")
```
