```python
"""
My Budget Tracker Prompt:
Create a Python program that asks the user for their monthly income and 3 expenses.
It should calculate the total expenses and remaining balance.
If the balance is negative, display a warning message.
"""
```

```python
income = float(input("Enter your income: "))

exp1 = float(input("Enter expense 1: "))
exp2 = float(input("Enter expense 2: "))
exp3 = float(input("Enter expense 3: "))

total = exp1 + exp2 + exp3
remaining = income - total
percentage = (total / income) * 100   # ← this is what you were adding

print("\n📊 Budget Summary:")
print(f"Total expenses: ${total:.2f}")
print(f"Remaining balance: ${remaining:.2f}")
print(f"Percentage spent: {percentage:.2f}%")   # ← this too

if remaining < 0:
    print("⚠️ Warning: You are over budget")
else:
    print("✅ You are within budget")
```
